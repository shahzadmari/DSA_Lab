```python
#store data
import numpy as np
import pandas as pd
#visualize
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

diamond = pd.read_csv("/content/Diamonds Prices2022.csv")


# Check the dimensions of our dataset
print(diamond.shape)

# Print the first 5 rows of the dataset
diamond.head()

# Find the total number of null or NaN values, in this case, there are none
diamond.isnull().sum()

#Categorize the values by a category, i.e cut in this case
diamond['cut'].value_counts()







```

```python

#generation_id are categorical
#let's plot it
sb.countplot(data=diamond, x='cut');
# ; supress the object information
```

```python

sb.color_palette()
base_color=sb.color_palette()[0]
sb.countplot(data=diamond, x='cut', color=base_color);
```

```python


color=diamond['color'].value_counts().index
color
sb.countplot(data=diamond, x='color', color=base_color, order=color);
```

```python


cut=diamond['cut'].value_counts().index
sb.countplot(data=diamond, y='cut', color=base_color, order=cut)
```

```python

# code for the pie chart generation_id above bar chart
sorted_counts = diamond['cut'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

```python

plt.hist(data=diamond, x='cut');
```

```python

sb.displot(diamond['cut']);

#showing different color bands with indexing first band has 0 index
```

```python

bins=np.arange(0, diamond['x'].max()+0.5, 0.5) # start, end, and step size
plt.hist(data=diamond, x='x', bins=bins);
```
